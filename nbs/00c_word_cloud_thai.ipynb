ไปดาวน์โหลด Font ที่ต้องการใช้ ใส่ไว้ใน path ที่กำหนด

```
path = '/content/THSarabunNew.ttf'
```

# 2. สร้าง Word Cloud

สร้าง mask ตามรูปร่างที่ต้องการ สีดำ พื้นขาว

```
mask = np.array(Image.open("/content/mask.png"))
```

ปรับ Parameter ได้ตามต้องการ

```
wordcloud = WordCloud(
                      font_path=path, 
                      relative_scaling = 0.3,
                      min_font_size=1,
                      background_color="white",
                      width=1024,
                      height=768,
                      max_words=2000, 
                      mask=mask,
                      colormap='plasma', 
                      scale=3,
                      font_step=4,
                    #   contour_width=3, 
                    #   contour_color='steelblue', 
                      collocations=False,
                      regexp=regexp,
                      margin=2
                      ).generate(text)
```

# 3. แสดงรูป Word Cloud ภาษาไทย

```
fig, ax = plt.subplots(1, 1, figsize=(16, 12))
ax.imshow(wordcloud, interpolation='bilinear')
ax.axis("off")
fig.show()
```
