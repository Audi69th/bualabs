# 0. Import

```
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

# 1. กำหนดคำตัวอย่าง

```
regexp = r"[ก-๙a-zA-Z']+"
```

```
text = "สักวา หวาน อื่น มี หมื่น แสน ไม่ เหมือนแม้น พจมาน ที่ หวาน หอม กลิ่นประเทียบ เปรียบดวง พวงพะยอม อาจ จะ น้อม จิต โน้ม"
```

ไปดาวน์โหลด Font ที่ต้องการใช้ ใส่ไว้ใน path

```
path = '/content/THSarabunNew.ttf'
```

# 2. สร้าง Word Cloud

```
wordcloud = WordCloud(
                      font_path=path, 
                      relative_scaling = 0.3,
                      min_font_size=1,
                      background_color="white",
                      width=1024,
                      height=768,
                      colormap='hsv', 
                      scale=3,
                      font_step=6,
                      collocations=False,
                      regexp=regexp,
                      margin=2
                      ).generate(text)
```

# 3. แสดงรูป Word Cloud ภาษาไทย

```
fig, ax = plt.subplots(1, 1, figsize=(16, 12))
ax.imshow(wordcloud, interpolation='bilinear')
ax.axis("off")
fig.show()
```
