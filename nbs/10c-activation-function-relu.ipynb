```python
import torch
import matplotlib.pyplot as plt
```

สร้าง x เป็นเลข ระหว่าง -10 บวกทีละ 0.01 ไปจนถึง 10 เราจะเอามาเป็นแกน x 

```python
x = torch.arange(-10., 10., 0.01)
```

```python
x
```

นำ x ผ่าน activation ฟังก์ชัน ได้ค่า y

```python
activation = torch.nn.Sigmoid()
y1 = activation(x)
```

```python
y1
```

```python
activation = torch.nn.ReLU()
y3 = activation(x)
```

```python
y3
```

แสดงกราฟ

```python
plt.ylim(top=2.0, bottom=-0.5)
plt.xlim(right=2.0, left=-2.0)
plt.legend(['Sigmoid', 'ReLU'], loc='upper left')
plt.grid()

plt.plot(x.numpy(), y1.numpy())
plt.plot(x.numpy(), y3.numpy())
plt.show()
```

# Leaky ReLU

ด้วยค่า negative_slope = 0.05

```python
activation = torch.nn.LeakyReLU(0.05)
y4 = activation(x)
```

แสดงกราฟ

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,6))
fig.suptitle('Comparing ReLU and LeakyReLU')
ax1.set_xlim([-3, 3])
ax1.set_ylim([-1, 3])
ax1.grid()
ax1.set_title('ReLU')
ax2.set_xlim([-3, 3])
ax2.set_ylim([-1, 3])
ax2.grid()
ax2.set_title('LeakyReLU')

ax1.plot(x.numpy(), y3.numpy())
ax2.plot(x.numpy(), y4.numpy())
```
