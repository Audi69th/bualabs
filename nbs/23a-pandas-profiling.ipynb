```
import numpy as np
import pandas as pd
import pandas_profiling

df = pd.DataFrame(
    np.random.rand(100, 5),
    columns=['a', 'b', 'c', 'd', 'e']
)
```

```
profile = pandas_profiling.ProfileReport(df, title="Test Title1")
profile
```

```
rejected_variables = profile.get_rejected_variables(threshold=0.9)
rejected_variables
```

```
profile.to_file("output.html")
```
